- Ch 9 Dictionaries: one of common use of dictionary is to count number of words in a text file
  ```python
   my_counter = dict()
   f_handle = open("a_text.text")
   for line in f_handle :
       line = line.rstrip()
       words = line.split()
       for w in words :
           my_counter[w] = my_counter.get(w, 0) + 1
   print(my_counter)
  ```
