# List of most useful Python package

## General-Purpose Utilities

- numpy – Fast numerical computing with arrays.
- pandas – Data manipulation and analysis.
- matplotlib – Basic plotting and visualization.
- seaborn – Advanced statistical visualizations.
- scipy – Scientific computing (optimization, integration, etc.).

## Machine Learning & AI

- scikit-learn – Classic machine learning models.
- tensorflow – Deep learning.
- torch – PyTorch, an alternative to TensorFlow.
- xgboost – Gradient boosting (for structured data).
- huggingface/transformers – NLP and large language models.

## Data Science & Statistics

- statsmodels – Statistical models (regression, time series).
- sympy – Symbolic mathematics.
- networkx – Graphs and network analysis.

## Web Scraping & Automation

- requests – HTTP requests.
- beautifulsoup4 – Web scraping.
- selenium – Automating browsers.
- scrapy – Advanced web scraping framework

## Data Visualization & Dashboards

- plotly – Interactive plots.
- bokeh – Web-based visualizations.
- altair – Declarative statistical visualization.
- dash – Web apps for data visualization.

## Web Development & APIs

- flask – Lightweight web framework.
- fastapi – Fast APIs with automatic documentation.
- django – Full-featured web framework.

## File Handling & System Tasks

- os – Interacting with the operating system.
- shutil – File operations.
- pathlib – Modern file path handling.
- pillow – Image processing.

## Performance & Parallel Computing

- numba – JIT compilation for speed.
- multiprocessing – Parallel computing.
- joblib – Parallelization made easy.

## Security & Cryptography

- cryptography – Encryption and security.
- hashlib – Hashing algorithms.

## Bonus: AI & Chatbots

- openai – OpenAI's API for ChatGPT and other models.
- langchain – Building AI-powered applications.

## How to install a package

`!pip install package_name`

## How to check a package is available

`!pip show package_name`

## How to show all installed packages

`!pip list`
